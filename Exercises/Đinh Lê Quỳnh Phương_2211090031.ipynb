```python
# Nhập thư viện
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu
x = np.array([5, 6, 8, 6, 3, 16, 2, 9, 3, 11, 12, 9, 6]).reshape(-1, 1)  # Số năm hút thuốc lá
y = np.array([99, 88, 89, 86, 111, 86, 104, 98, 64, 81, 82, 84, 86])  # Chỉ số CEA

# Vẽ dữ liệu ban đầu
plt.scatter(x, y, color='red', label='Dữ liệu ban đầu')
plt.xlabel('Số năm hút thuốc lá')
plt.ylabel('Chỉ số CEA')

# Tạo và huấn luyện mô hình
model = LinearRegression()
model.fit(x, y)

# Đường hồi quy
y_pred = model.predict(x)
plt.plot(x, y_pred, color='blue', label='Đường hồi quy')
plt.legend()
plt.show()

# Dự đoán khi x = 10
x_predict = np.array([[10]])
y_predict = model.predict(x_predict)
print(f"Phương trình hồi quy: y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x")
print(f"Dự đoán chỉ số CEA khi x = 10: y = {y_predict[0]:.2f}")
```
