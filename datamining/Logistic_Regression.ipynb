```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.metrics import accuracy_score
samples=datasets.load_breast_cancer()
#print(samples.DESCR)
#M: Ác tính (malignant)
#B: Lành tính (benign)
x=samples.data[:,0:6:5]
y=samples.target
M_set=np.array([x[i] for i in range(len(y)) if y[i]==0])
B_set=np.array([x[i] for i in range(len(y)) if y[i]==1])
plt.xlabel("Bán kính")
plt.ylabel("Độ lõm")
plt.scatter(M_set[:,0], M_set[:,1],color='red')
plt.scatter(B_set[:,0], B_set[:,1],color='blue')
logRegr=LogisticRegression()
logRegr.fit(x,y)
y_pred=logRegr.predict(x)

M_set=np.array([x[i] for i in range(len(y)) if y_pred[i]==0])
B_set=np.array([x[i] for i in range(len(y)) if y_pred[i]==1])
plt.xlabel("Bán kính")
plt.ylabel("Độ lõm")
plt.scatter(M_set[:,0],M_set[:,1],color='red')
plt.scatter(B_set[:,0],B_set[:,1],color='blue')
needPrediction=[[16,0.03],[15,0.02],[11,0.18]]
preds=logRegr.predict(needPrediction)
for pred in preds:
  if pred==0:
    print("Ác tính")
  else:
    print("Lành tính")
print(accuracy_score(y,y_pred))
```
