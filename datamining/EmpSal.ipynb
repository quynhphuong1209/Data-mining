```python
from google.colab import drive
drive.mount('/content/drive')

#Import libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
os.chdir('/content/drive/My Drive/Datamining/Exercises')

#Load data-set
dataset=pd.read_csv('EmpSal.csv')
dataset.describe()
dataset.corr()
#Drop ID column
df=dataset.drop(['ID'],axis=1)

#Bieu do rau meo (boxplot)
df.plot(kind='box')
plt.title('Unprocessing data')
plt.show()
print('\n')

#Method 1: Standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
dataStand=sc.fit_transform(df)

#Bieu do rau meo (boxplot)
df.plot(kind='box')
plt.title('Unprecessing data')
plt.show()
print('\n')

#Method 2: Normalization
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
dataNorm=mn.fit_transform(df)

dataNorm=pd.DataFrame(dataNorm,columns=['Experience_Years','Salary'
])
#Replotting after Normalization
dataNorm.plot(kind='box')
plt.title('After Normalization')
plt.show()
print('\n')
dataNorm.plot(kind='scatter',x='Experience_Years',y='Salary')
plt.show()
```
